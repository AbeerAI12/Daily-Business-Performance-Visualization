```python
# استيراد المكتبات
import pandas as pd
import matplotlib.pyplot as plt

# إنشاء البيانات (عدد الأيام، المبيعات، التكاليف)
data = {
    "Day": [1, 2, 3, 4, 5],
    "Sales": [1200, 1500, 1100, 1800, 1400],
    "Costs": [800, 1000, 1200, 1300, 1500]
}

# تحويل البيانات إلى DataFrame
df = pd.DataFrame(data)

# حساب الربح أو الخسارة
df["Profit_Loss"] = df["Sales"] - df["Costs"]

# عرض الجدول
print(df)

# رسم المبيعات والتكاليف
plt.figure()
plt.plot(df["Day"], df["Sales"], label="Sales")
plt.plot(df["Day"], df["Costs"], label="Costs")
plt.xlabel("Day")
plt.ylabel("Amount")
plt.title("Daily Sales and Costs")
plt.legend()
plt.show()

# رسم الربح والخسارة
plt.figure()
plt.bar(df["Day"], df["Profit_Loss"])
plt.xlabel("Day")
plt.ylabel("Profit / Loss")
plt.title("Daily Profit and Loss")
plt.show()
```
